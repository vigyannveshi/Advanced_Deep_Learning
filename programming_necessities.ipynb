1) **`np.meshgrid`**<br>
It is used to create a rectangular grid - two "one dimensional" arrays.

```python
import numpy as np
x=np.arange(5)
y=np.arange(10)
X,Y=np.meshgrid(x,y)
```

**3D plotting using meshgrid**

```python
import matplotlib.pyplot as plt
import matplotlib_inline
matplotlib_inline.backend_inline.set_matplotlib_formats('svg')
fig=plt.figure()
ax=fig.add_subplot(111,projection="3d")

x=np.arange(-50,50)
y=np.arange(-50,100)

X,Y=np.meshgrid(x,y)

Z=np.sqrt((X**2+Y**2)) 

ax.plot_surface(X,Y,Z)
plt.show()
```

3) **`np.unravel_index(i,(a,b))`**<br>
It is used to get the ith index from a indiced list


```python
import numpy as np

for i in range(24):
    print(np.unravel_index(i,(4,6)))
```
